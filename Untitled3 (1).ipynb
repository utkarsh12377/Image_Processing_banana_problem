```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import morphology, measure, segmentation, color, util
from skimage.morphology import skeletonize, thin, disk, convex_hull_image
from scipy.ndimage import binary_fill_holes
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from skimage.segmentation import slic, mark_boundaries

# Create an output folder
OUT_DIR = "/content/assignment_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

# Helper: save & show images
def save_im(path, img, cmap='gray'):
    plt.imsave(path, img if img.ndim == 3 else img, cmap=cmap)
    print("‚úÖ Saved:", path)
```

```python
# Set your uploaded file paths (they should exist in /content)
INPUT_PATH_1A = "/content/f3_a00090_jpg.rf.da42bd37335cd2feaa0534d4e6b55c46.jpg"
INPUT_PATH_1B = "/content/9ca97574cdfcbcd1db8ba2fb76e2dd34.png"
INPUT_PATH_2  = "/content/15350.jpg"

# Verify all files exist and provide a more helpful error message
import os
missing_files = [path for path in [INPUT_PATH_1A, INPUT_PATH_1B, INPUT_PATH_2] if not os.path.exists(path)]

if missing_files:
    error_message = "The following input files were not found:\n"
    for file in missing_files:
        error_message += f"- {file}\n"
    error_message += "\nPlease upload the missing files to the /content directory in the Colab environment."
    raise FileNotFoundError(error_message)

for path in [INPUT_PATH_1A, INPUT_PATH_1B, INPUT_PATH_2]:
    print(path, "‚úÖ" if os.path.exists(path) else "‚ùå not found!")
```

```python
print("=== Assignment 1A: Morphological Image Processing ===")

# Load the damaged fabric image
img_fabric = cv2.imread(INPUT_PATH_1A, cv2.IMREAD_GRAYSCALE)
if img_fabric is None:
    raise FileNotFoundError("Fabric image not found! Check file path or upload again.")

# Convert to binary (white = threads, black = defects)
# Adjust threshold value (100‚Äì150) for your image contrast
_, fabric_bin = cv2.threshold(img_fabric, 120, 255, cv2.THRESH_BINARY)
fabric_bin = fabric_bin // 255  # normalize to 0/1 for morphology

# Create structuring element (you can tune disk radius)
from skimage.morphology import binary_dilation, binary_erosion, opening, closing
selem = disk(3)

# --- 1Ô∏è‚É£ Dilation ---
dilation = binary_dilation(fabric_bin, selem)

# --- 2Ô∏è‚É£ Erosion ---
erosion = binary_erosion(fabric_bin, selem)

# --- 3Ô∏è‚É£ Opening (erosion ‚Üí dilation) ---
opening_res = opening(fabric_bin, selem)

# --- 4Ô∏è‚É£ Closing (dilation ‚Üí erosion) ---
closing_res = closing(fabric_bin, selem)

# --- 5Ô∏è‚É£ Top-hat Transformation ---
# Highlights small bright elements (thread patterns)
tophat = fabric_bin - opening_res
tophat = (tophat > 0).astype(np.uint8)

# --- 6Ô∏è‚É£ Black-hat Transformation ---
# Highlights small dark regions (holes/defects)
blackhat = closing_res - fabric_bin
blackhat = (blackhat > 0).astype(np.uint8)

# --- 7Ô∏è‚É£ Skeletonization ---
skeleton = skeletonize(fabric_bin)

# --- 8Ô∏è‚É£ Thinning ---
thinning = thin(fabric_bin)

# --- 9Ô∏è‚É£ Thickening ---
thickening = binary_dilation(thinning, disk(1))

# --- üîü Convex Hull Extraction ---
labels = measure.label(fabric_bin)
hull_img = np.zeros_like(fabric_bin)
for region in measure.regionprops(labels):
    if region.area < 10:
        continue
    minr, minc, maxr, maxc = region.bbox
    patch = (labels[minr:maxr, minc:maxc] == region.label)
    hull = convex_hull_image(patch)
    hull_img[minr:maxr, minc:maxc] |= hull

# --- 1Ô∏è‚É£1Ô∏è‚É£ Hole Filling ---
filled = binary_fill_holes(fabric_bin)

# --- 1Ô∏è‚É£2Ô∏è‚É£ Boundary Extraction ---
# Extracts object boundaries (white edges of fabric threads)
boundary = fabric_bin - binary_erosion(fabric_bin, disk(1))
boundary = (boundary > 0).astype(np.uint8)

# --- üíæ Save all outputs ---
morph_outputs = {
    "input_binary": fabric_bin,
    "dilation": dilation,
    "erosion": erosion,
    "opening": opening_res,
    "closing": closing_res,
    "tophat": tophat,
    "blackhat": blackhat,
    "skeleton": skeleton,
    "thinning": thinning,
    "thickening": thickening,
    "convex_hull": hull_img,
    "hole_filling": filled,
    "boundary": boundary
}

for name, img in morph_outputs.items():
    save_im(os.path.join(OUT_DIR, f"1A_{name}.png"), img, cmap='gray')

print("\n‚úÖ All 12 morphological operations completed successfully!")
print("üìÇ Check output folder:", OUT_DIR)
```

```python
print("=== Assignment 1B: Chest X-ray ===")

img_xray = cv2.imread(INPUT_PATH_1B, cv2.IMREAD_GRAYSCALE)
if img_xray is None:
    raise FileNotFoundError("X-ray not found!")

# 1Ô∏è‚É£ Canny Edge Detection
edges = cv2.Canny(img_xray, 80, 180)
save_im(os.path.join(OUT_DIR, "1B_canny.png"), edges)

# 2Ô∏è‚É£ K-means Segmentation
X = img_xray.reshape(-1, 1).astype(np.float32)
kmeans = KMeans(n_clusters=3, random_state=0, n_init=10)
labels = kmeans.fit_predict(X)
seg = labels.reshape(img_xray.shape)
save_im(os.path.join(OUT_DIR, "1B_kmeans.png"), seg)

# 3Ô∏è‚É£ SLIC Superpixels
segments = slic(util.img_as_float(img_xray), n_segments=400, compactness=10, sigma=1, start_label=1, channel_axis=None)
boundaries = mark_boundaries(color.gray2rgb(util.img_as_float(img_xray)), segments)
save_im(os.path.join(OUT_DIR, "1B_slic.png"), boundaries)

print("‚úÖ Assignment 1B completed! Check folder:", OUT_DIR)
```

```python
print("=== Assignment 2: Feature Detection & PCA ===")

img_leaf = cv2.imread(INPUT_PATH_2)
img_gray = cv2.cvtColor(img_leaf, cv2.COLOR_BGR2GRAY)

# MSER
mser = cv2.MSER_create(min_area=60, max_area=8000)
regions, _ = mser.detectRegions(img_gray)
mser_vis = img_leaf.copy()
for p in regions:
    hull = cv2.convexHull(p.reshape(-1,1,2))
    cv2.polylines(mser_vis, [hull], True, (0,255,0), 1)
save_im(os.path.join(OUT_DIR, "2_mser.png"), cv2.cvtColor(mser_vis, cv2.COLOR_BGR2RGB))

# Harris Corner
dst = cv2.cornerHarris(np.float32(img_gray), 2, 3, 0.04)
corner_img = img_leaf.copy()
corner_img[dst > 0.01*dst.max()] = [0,0,255]
save_im(os.path.join(OUT_DIR, "2_harris.png"), cv2.cvtColor(corner_img, cv2.COLOR_BGR2RGB))

# SIFT (requires opencv-contrib)
sift = cv2.SIFT_create()
kp = sift.detect(img_gray, None)
img_sift = cv2.drawKeypoints(img_leaf, kp, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
save_im(os.path.join(OUT_DIR, "2_sift.png"), cv2.cvtColor(img_sift, cv2.COLOR_BGR2RGB))

# PCA Reconstruction
h, w = img_gray.shape
X = img_gray / 255.0
pca = PCA(n_components=50)
X_proj = pca.fit_transform(X)
X_recon = pca.inverse_transform(X_proj)
recon = np.clip(X_recon*255, 0, 255).astype(np.uint8)
save_im(os.path.join(OUT_DIR, "2_pca.png"), recon, cmap='gray')

print("‚úÖ Assignment 2 completed! Outputs saved in:", OUT_DIR)
```
